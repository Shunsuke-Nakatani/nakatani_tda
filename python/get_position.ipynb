```python
##sample.csvの読み取り
import numpy as nm
import pandas as pd
import os
os.chdir("/Users/sinco/OneDrive/デスクトップ/python")

df =pd.read_csv('sample.csv')
#sample.csvからfrom列とto列を抽出(listではない)
fr =df["from_address"]
to =df["to_address"]

#from列だけのデータ
df1 = pd.read_csv('sample.csv')
df1.drop(columns=["value","to_address","block_timestamp","transaction_index"],inplace = True)

#to列だけのデータ
df2 = pd.read_csv('sample.csv')
df2.drop(columns=["value","from_address","block_timestamp","transaction_index"],inplace = True)

print(df2)


#from列、to列をリスト化、frliの要素数が22個になってしまう
frli = fr.values.tolist()
frli1 = fr.values.tolist()
toli = to.values.tolist()

#from,toにあるすべてのアカウント名のリスト(順不同)
al = frli
al.extend(toli)
al1=[]
al1=list(set(al))

 #print(len(frli1),len(toli))
```

```python
#グラフの表示
import py4cytoscape as p4c
nodes = pd.DataFrame(data={'id':al1})
edges = pd.DataFrame(data={'source':frli1,'target':toli})
p4c.create_network_from_data_frames(nodes, edges, title="my first network", collection="DataFrame Example")
```

```python
#ノードの座標の取得から表の作成(from)
n = p4c.get_node_position()
n['index'] = n.index
n.columns = ['x','y','from_address']
n['x'] = n['x']/4
n['y'] = n['y']/4
print(n)
```

```python
#fromの座標を表に
fr_coords = pd.merge(df1,n)
print(fr_coords)

fr_coords1 = pd.merge(df1,n)
fr_coords1.drop(columns=["from_address"],inplace = True)
print(fr_coords1)
```

```python
#fromアドレスのリスト化
coordinates1 = [list(x) for x in fr_coords1.values]
print(coordinates1)
```

```python
#ノードの座標の取得から表の作成(to)
m = p4c.get_node_position()
m['index'] = m.index
m.columns = ['x','y','to_address']
m['x'] = m['x']/4
m['y'] = m['y']/4
print(m)
```

```python
#toの座標を表に
to_coords = pd.merge(df2,m, how ='left')
print(to_coords)

fr_coords2 = pd.merge(df2,m, on = 'to_address', how= 'left')
fr_coords2.drop(columns=["to_address"],inplace = True)
print(fr_coords2)
```

```python
#toアドレスのリスト化
coordinates2 = [list(y) for y in fr_coords2.values]
print(coordinates2)
```

```python
#Wirecloudのimportdataフォームに変換
import copy

entities=[]
dat ={}

for i,acc in enumerate(fr):
    dat["id"]="urn:ngsi-ld:Transcation:%d" %i    #idつける

    dat["type"]="Transaction"    #typeつける

    dat["name"]={"type":"Text","value":acc}   #nameつける
      #locationつける
    value={"type":"LineString","coordinates":[coordinates1[i],coordinates2[i]]}
    dat["location"]={"type":"geo:json","value":value}

    entities.append(copy.deepcopy(dat))

print(entities)
```

```python
#json形式に書き換える(entityデータの出力)
import json
entities_json=json.dumps(entities,indent=3)
print(entities_json)
```
